```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as psf
from pyspark.sql.functions import col, split, lit, when, udf
from pyspark.sql.types import *
from datetime import datetime
```

```python
spark = SparkSession.builder.appName('files').getOrCreate()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.dpi"] = 144
import pandas as pd
%load_ext sql_magic
%config SQL.conn_name = 'spark'
```

```python
df = spark.read.json('data/files/', multiLine=True)
```

```python
df.show()
```

```python
explodedDF = df.select(psf.explode(df.data))
explodedDF.take(1)
```

```python
objectDF = explodedDF.select(explodedDF["col"].getItem('fileMode').alias('fileMode'),
                             explodedDF["col"].getItem('path').alias('path'),
                             explodedDF["col"].getItem('filename').alias('filename'),
                             explodedDF["col"].getItem('size').alias('size'),
                             explodedDF["col"].getItem('lastModified').alias('lastModified'))
```

```python
objectDF.show(truncate=False)
```

```python
objectDF_PD = objectDF.toPandas()
objectDF_PD.plot.bar(x='filename',y='size')
```
